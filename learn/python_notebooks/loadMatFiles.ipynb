## Loading files from matlab
This example show how simple is to load data from matlab files.

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load mat file, will return a dictionary where every entry will be a variable
# from the mat file
dictMat = scipy.io.loadmat('./simpleMatFile.mat')

# Select variables/matrices
matA = dictMat['matA']
matB = dictMat['matB']
simpleVec_y = dictMat['y']
simpleVec_x = dictMat['x']
```

### Shapes
Observe that you don't need to care about the col/row major order. The matrix shapes will be the same.

```python
print("Shape of matrix A", matA.shape)
print("Shape of matrix B", matB.shape)
print("\n Matrix B")
plt.matshow(matB); plt.show()
print(matB)
print("\n Matrix A")
plt.matshow(matA); plt.show()
print(matA)
```

### Displaying vector
The only difference that you must pay attention is the vector, on matlab [1,1001] need to be transposed to (1001)

```python
print("Shape simpleVec_y", simpleVec_y.shape)
plt.plot(simpleVec_y.reshape([np.prod(simpleVec_y.shape)]))
plt.show()
```

### Saving data to matfiles
We can also save data to matlab.

```python
matNumpy_2x3 = np.matrix('1 2 3; 4 5 6')
print(matNumpy_2x3)
dictSaveMat={}
dictSaveMat['Matrix_2x3_from_numpy']=matNumpy_2x3.astype('float')
scipy.io.savemat('matFromPython',dictSaveMat)
```
